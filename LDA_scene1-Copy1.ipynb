```python
def preprocessing(doc):
    morphs = []
    noun_adj_list=[]
    stopwords = ['하는','남', '감', '욕', '항', '움', '함', '스러', '계심','억', '중', '대감', '수','줄', '심', '로움','간', '있다', '있는',
             '상황', '모습', '사진','중이', '것','하고','같다', '하러', '해서','대해','있음', '할', '않아', '된', '뭔가', '이', '그', '못', 
              '해', '않은', '하며', '있는데','만','있어', '이기', '하고있다', '있지만', '듯', '스럽고', '했는데','않고', '하여', '되지','스럽다',
             '하다', '이를','하던','하면서','하려','고', '싶다','상치','하지','를','않는','거','치', '같음']
    for word in doc:
        morphs.append(okt.pos(word))
    for words in morphs:
        for word, tag in words:
            if tag in ['Noun','Adjective','Verb']:
                if word not in stopwords:
                    noun_adj_list.append(word)
    return noun_adj_list
    
```

```python
doc1 = preprocessing(categ['1'])
doc2 = preprocessing(categ['2'])
doc3 = preprocessing(categ['3'])
doc4 = preprocessing(categ['4'])
doc5 = preprocessing(categ['5'])
doc6 = preprocessing(categ['6'])
doc7 = preprocessing(categ['7'])
doc8 = preprocessing(categ['8'])
doc25 = preprocessing(categ['25'])
doc26= preprocessing(categ['26'])
doc27 = preprocessing(categ['27'])
doc28 = preprocessing(categ['28'])
doc29 = preprocessing(categ['29'])
doc30 = preprocessing(categ['30'])
doc31 = preprocessing(categ['31'])
doc32 = preprocessing(categ['32'])
```

```python
tokenized_data = [doc1, doc2, doc3, doc4, doc5, doc6, doc7, doc8, doc25, doc26, doc27, doc28, doc29, doc30, doc31, doc32]
print(tokenized_data)
```

```python
dictionary = corpora.Dictionary(tokenized_data)
```

```python
corpus = [dictionary.doc2bow(text) for text in tokenized_data]
print(corpus[4])
```

```python
lda_model = models.LdaModel(corpus=corpus, num_topics=NUM_TOPICS, id2word=dictionary)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X = vectorizer.fit_transform(detok[0])
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation(n_components=15, learning_method='online', random_state=777, max_iter=1)
lda_top = lda_model.fit_transform(X)
print(lda_model.components_)
print(lda_model.components_.shape)
```

```python
terms = vectorizer.get_feature_names() 
# 단어 집합. 1,000개의 단어가 저장되어있음.
def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d :" % (idx+1), [(feature_names[i], topic[i].round(2)) for i in topic.argsort()[:-n -1:-1]])
get_topics(lda_model.components_, terms)
```
